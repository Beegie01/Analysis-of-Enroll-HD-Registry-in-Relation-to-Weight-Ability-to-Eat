```python
fname = 'enrl_per_vis.csv'
all_vis = pd.read_csv(data_dir+fname)
display(all_vis.info(), all_vis)
```

```python
fname = 'only6_visits.csv'
first6_vis = pd.read_csv(data_dir+fname)
display(first6_vis.info(), first6_vis)
```

```python
ctl =  (first6_vis['hd_status0'].isin(['genotype -ve']))

ctlm = ((first6_vis['hd_status0'].isin(['genotype -ve'])) &
       (first6_vis['is_male'] == 1))

ctlf = ((first6_vis['hd_status0'].isin(['genotype -ve'])) &
       (first6_vis['is_male'] == 0))
       

control = first6_vis.loc[ctl].dropna(axis=1)

male_control = first6_vis.loc[ctlm].dropna(axis=1)

female_control = first6_vis.loc[ctlf].dropna(axis=1)

display(control['visitnum'].value_counts(), control, male_control, female_control)
```

```python
mnf =  (first6_vis['hd_status0'] == 'manifest')

mnfm = ((first6_vis['hd_status0'] == 'manifest') &
       (first6_vis['is_male'] == 1))

mnff = ((first6_vis['hd_status0'] == 'manifest') &
       (first6_vis['is_male'] == 0))
       

manifest = first6_vis.loc[mnf].dropna(axis=1)

male_manifest = first6_vis.loc[mnfm].dropna(axis=1)

female_manifest = first6_vis.loc[mnff].dropna(axis=1)

display(manifest['visitnum'].value_counts(), manifest, male_manifest, female_manifest)
```

### VARIABLE ANALYSIS PER  VISIT

#### Independence Score

```python
ctrl_avg_indepscl = pd.Series(control.xs(('indepscl', ), axis=1).mean().round(1), name='control')
mnf_avg_indepscl = pd.Series(manifest.xs(('indepscl', ), axis=1).mean().round(1), name='manifest')
display(ctrl_avg_indepscl, mnf_avg_indepscl)
avg_indepscl = pd.concat([ctrl_avg_indepscl, mnf_avg_indepscl], axis=1)
display(avg_indepscl)
fig, ax1 = plt.subplots(figsize=[4, 4], dpi=200)
proj.adjust_axis(avg_indepscl.plot(kind='bar', ax=ax1), 
                 plot_title='Average Independence Score per Visit',
                 x_labe='Visit No', y_labe='average indepscl', annotate=True, annot_size=5)
sns.move_legend(ax1, [0.25, 0.4])
```

```python
proj.report_a_significance(avg_indepscl['control'], avg_indepscl['manifest'], X1_name='control', X2_name='manifest')
```

```python
ctrl_avg_miscore = pd.Series(control.xs(('miscore', ), axis=1).mean().round(1), name='control')
mnf_avg_miscore = pd.Series(manifest.xs(('miscore', ), axis=1).mean().round(1), name='manifest')
display(ctrl_avg_miscore, mnf_avg_miscore)
avg_miscore = pd.concat([ctrl_avg_miscore, mnf_avg_miscore], axis=1)
display(avg_miscore)
fig, ax1 = plt.subplots(figsize=[4, 4], dpi=200)
proj.adjust_axis(avg_miscore.plot(kind='bar', ax=ax1), 
                 plot_title='Average Miscore per Visit',
                 x_labe='Visit No', y_labe='average miscore', annotate=True, annot_size=5)
sns.move_legend(ax1, [0.25, 0.4])
```

```python
proj.report_a_significance(avg_miscore['manifest'], avg_miscore['control'], X2_name='control', X1_name='manifest')
```
